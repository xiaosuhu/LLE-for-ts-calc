```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Folder containing the CSV files
folder_path = "./roi-ts-data"

# Track files with missing data
files_with_missing_data = []
```

```python

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        file_path = os.path.join(folder_path, filename)
        # df = pd.read_csv(file_path)
        data = np.loadtxt(file_path)
        df = pd.DataFrame(data)

        if df.isnull().values.any():
            print(f"Missing data found in: {filename}")
            files_with_missing_data.append(filename)

if not files_with_missing_data:
    print("No missing data found in any file.")
```
