```python
# Folder containing CSV files
data_folder = "./roi-ts-data"
files = sorted([f for f in os.listdir(data_folder) if f.endswith(".csv")])
fs = 1/0.8  # sampling rate in Hz (change this for your data)
```

```python
# Preallocate results list
results = []
filenames = []

# Loop through each file and compute LLE for each channel
for filename in files:
    file_path = os.path.join(data_folder, filename)
    data = np.loadtxt(file_path)
    # df = pd.read_csv(file_path)

    lle_row = []
    for ch in range(4):
        signal = data[:, ch]
        try:
            lle = estimate_LLE_auto(signal,fs=fs)
        except Exception as e:
            print(f"Error in file {filename}, channel {ch}: {e}")
            lle = np.nan
        lle_row.append(lle)

    results.append(lle_row)
    filenames.append(os.path.splitext(filename)[0])  # Remove .csv extension

# Create DataFrame with filenames as index
lle_df = pd.DataFrame(results, columns=[f"ROI_{i+1}" for i in range(4)], index=filenames)

# Save to CSV
lle_df.to_csv("lle_results.csv")

print("LLE computation complete. Result saved to lle_results.csv.")
```
